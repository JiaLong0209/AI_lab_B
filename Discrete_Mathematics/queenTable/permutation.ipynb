```python
# Solution 1. 自己想的解法, Recursion and for loop
def all_permutations(input_list):

    def permute(p_list = [], index = 0):
        if len(p_list) == len(input_list):         # Condition for ending this function
            total.append(p_list)
            return

        for i in range(len(input_list)):             # Call itself using each index of the list
            new_list = [*p_list, input_list[i]]      # Add a new value to the back
            if len(set(new_list)) == len(new_list):  # Check if the list has same value
                permute(new_list, i)

    total = []
    permute()   # Call permute function
    return total

n = 5
input_list = list(range(1, n+1))
total = all_permutations(input_list)

print(total)
print(len(total))

for i, v in enumerate(total):
    print(f'{i+1}. {v}')
```

```python
# Solution 2. ChatGPT
def get_permutations(lst):
    def generate_permutations(current, remaining):
        if not remaining:
            permutations.append(current[:])
            return
        for i in range(len(remaining)):
            current.append(remaining[i])
            generate_permutations(current, remaining[:i] + remaining[i+1:])
            current.pop()

    permutations = []
    generate_permutations([], lst)
    return permutations

# Example usage:
n = 5
input_list = list(range(1, n+1))
result = get_permutations(input_list)
print(result)
print(len(result))
```

```python
# Solution 3. 上網找的解法, reduce
from functools import reduce

n = 5
num_list = list(range(1, n+1))

def permute(nums):
    return [[int(x) for x in list(i)] for i in reduce(lambda acc,cur:[str(a)+str(b) for a in acc for b in cur if str(b) not in str(a) ],[nums]*len(nums))]

print(permute(num_list))
print(len(permute(num_list)))
```
