```python
it_url = 'https://ithelp.ithome.com.tw/'
path = ''
headers = {
   "User-Agent" : 'Mozilla/5.0 (X11; Linux x86_64; rv:124.0) Gecko/20100101 Firefox/124.0'
}
```

```python
resp = requests.get(it_url, headers = headers)
# resp = requests.get(it_url)
status_code = resp.status_code


print(status_code)
# print(resp.content) # binary content
# print(resp.text) # html text
# print(resp.content.decode())  # content to string
html = etree.HTML(resp.text) # html node tree
print(html)
```

```python
titles = html.xpath("//div[@class='qa-list'][1]//a[@class='qa-list__title-link']/text()")
print(len(titles),titles)

```

```python
import os
def get(url, headers = None):
    return requests.get(url, headers = headers)

dir = './imgs/'
url = 'https://www.pinterest.jp/aitiziku/%E6%9D%B1%E6%96%B9project/'
resp = get(url, headers = headers)
html = etree.HTML(resp.content)
print(resp.status_code)
imgs = html.xpath('//div[@class="PinCard__imageWrapper"]//img/@src')
print(len(imgs),imgs)

if(os.path.isdir(dir) == False):
    os.mkdir(dir)

for i in imgs:
    file_name = i.split('/')[-1]
    file_content = requests.get(i, headers= headers).content
    print(file_content)
    with open(dir + file_name, "wb") as f:
        f.write(file_content)


```
