```python
cookies = {'over18': '1'}
web = requests.get(ptt_url, cookies = cookies)
soup = BeautifulSoup(web.text, "html.parser")

res = soup.select('.r-ent')
res_find = soup.find_all('div', class_='r-ent')

print(type(soup))
print(type(res))
print(type(res_find))

for i in res_find:
    a = i.find_all('a')
    for index,j in enumerate(a):
        if(index == 0):
            text = j.get_text()
        elif(index == 2):
            name = j.get_text()[6:-4]
    href = a[0]['href']
    print(name)
    print(text)
    print(ptt_host + href)
```

```python
web = requests.get(ptt_url, cookies=cookies)
soup = BeautifulSoup(web.text, 'html.parser')
cookies = {'over18': '1'}

terms = ['title', 'link', 'id', 'time']
content = {i: '' for i in terms}
articles = soup.find_all('div', class_='r-ent')
fun = [lambda i: i.find('a').get_text(), lambda i: ptt_host + i.find('a')['href'],  lambda i: i.find('div', class_="author").get_text(), lambda i: i.find('div', class_='date').get_text()]
page = []

for i in articles:
    for x, t in enumerate(terms):  content[t] = fun[x](i)
    content = {i: '' for i in terms}
    page.append(content)

for i in page:
    print(i)

with open('w2.json', 'w', encoding='UTF-8') as f:
    json.dump(page, f, ensure_ascii=False, indent=4)
```
