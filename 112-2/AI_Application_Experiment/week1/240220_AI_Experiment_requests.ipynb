```python
import requests as req

def check(url, headers = None):
    r = req.get(url, headers)
    print(f'Try to connect {url}')
    if r.status_code == 200:
        html_text = r.text
        print("Request successful")
    else:
        print("Request failed")
    print(r.status_code, req.codes.ok)
    print()


print(req)
url_tku = "https://www.tku.edu.tw/"
url_lvimWindows = "https://medium.com/@peterhaddad/how-to-setup-lunarvim-in-windows-9df9bd220889"
url_http_get = "http://httpbin.org/get"
url_http_post = "http://httpbin.org/post"
```

```python
check(url_tku)
check(url_lvimWindows)
```

```python
params = {"name": "lunarVim", "class": "Linux"}
r = req.get(url_http_get, params = params)
print(r.status_code)
print(r.text)
```

```python
print(r.status_code, req.codes.ok)
params = {"name": "lunarVim", "class": "Linux"}
r = req.post(url_http_post, params = params)
print(r.status_code)
print(r.text)
```

```python
r = req.get(url_tku)
ret = [r, r.text, r.content, r.raw]
for i in ret:
    print('-' * 190)
    print(i)
```

```python
r = req.get(url_http_get)
headers = ['Content-Type', 'Content-Length', 'Date', 'Server']
for i in headers:
    print(r.headers[i])
```
